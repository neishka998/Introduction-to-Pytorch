# About Tensors

*We would be learning about Tensors. Tensors can be a number, a metrix, a vector or a 3D array. In the follwoing example, using ```'.'``` indicates the data type to be a float.*

````python
# Creating our first variable T1. 
T1 = torch.tensor(3.)
T1
````

````python
# For more complex examples, Lets try creating a vector Tensor
T2 = torch.tensor([2, 1., 4, 6, 5.])
print('Vector Tensor:  \n' , T2)
````

````python
# For more complex examples, Lets try creating a matrix Tensor

T3 = torch.tensor([[2, 1.], 
                  [5, 7.], 
                  [10., 5 ]])
print('Matrix Tensor:  \n' , T3)
````

````python
# For more complex examples, Lets try creating a 3-D Array Tensor

T4 = torch.tensor([[[2, 1. , 4], 
                  [5, 7., 9]], 
                  [[10., 5 ,16] ,
                  [33, 56. , 66]]])
print('3-D Array:  \n' , T4)
````

# Tensor Functions

*The Torch module also contains many functions for creating and manipulating tensors. *

* *Creating a matrix of a size *3x2* *having a fixed value.*
* **Cat** *function joins the two tensors. Make sure the tensors you want to join should be of the same size.*
* *Using functions like ```tan, cos, sin```*
* *Changing the shape of an existing tensor / concatenated tensor.*

````python
T5 = torch.full((3,2), 42) #Using full fucntion to create a new tenor 
print('New Tensor T5:  \n' , T5) # Printing the value 
````

````python
# Joining two tenors with same dimmensions by using CAT fucntion
T6 = torch.cat((T3, T5))
print('T6:  \n' , T6) # Printing the value 
````

````python
# Using mathematical functions like tan, sin, cos

T7 = torch.sin(T6)
print('T7:  \n' , T7) # Printing the value 
````

````python
# Change the shape of a tensor
T8 = T7.reshape((3,2,2))
print('T8:  \n' , T8) # Printing the value 
````

# Using Numpy & Pandas

* *Converting a numpy array to a  torch tensor using* ```torch.from_numpy```

* *Converting a tensor to numpy array using* ``` .numpy() ```

````python
import numpy as np 

x = np.array([[2,4.5],[5,3],[7.2,1],[9,10]]) #Creating a numpy Array
print('This is a numpy array: \n' , x) 

tensor_arr = torch.from_numpy(x) # Converting the numpy array to tensor
print('Tensor Array:  \n' , tensor_arr)

#Checking the type of both the arrays
ty_x = x.dtype
ty_y = tensor_arr.dtype

print(ty_x,ty_y)
````

````python
#Converting tensor to numpy 
num_arr = tensor_arr.numpy()
print('This is the converted numpy array:  \n' , num_arr)
````

*Interoperability between PyTorch and Numpy is essential.* 

**WHY ?**

*Because the data reading and processing would be mostly done with the help of numpy.*

**Then why PyTorch at all ??**
* **Autograd**
The ability to ```automatically``` computing the ```gradients``` for tensor operation which is quite essential for deep learning models. 

* **GPU** 
As we would be working with a large dataset, operations can be perfomaed efficiently using GPU.
